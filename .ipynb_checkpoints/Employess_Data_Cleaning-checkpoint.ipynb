```python
import numpy as np
```

```python
import pandas as pd
df = pd.read_excel(r'employees.xlsx')
df
```

## Handling missing values

```python
# Replace 'NaN' with actual NaN values
df.replace('NaN', np.nan, inplace=True)

# Remove rows with missing values (NaN)
# df.dropna(inplace=True)

# Fill missing values with appropriate replacements
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Salary'] = df['Salary'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
df = df.reset_index(drop=True)
df
```

## Standardising data

```python
df['Gender'] = df['Gender'].str.upper()
df
```

## Remove duplicates

```python
df.drop_duplicates(inplace=True)
df
```

## Correcting inconsistent data

```python
df['Phone'] = df['Phone'].str.replace('[^0-9]', '', regex=True)
df
```

## Outlier Detection and Handling:


```python
# Identify and handle outliers in the 'Score' column (e.g., values above 100)
score_threshold = 100
df.loc[df['Score'] > score_threshold, 'Score'] = score_threshold
df
```
